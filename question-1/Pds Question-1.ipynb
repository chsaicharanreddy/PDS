```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df = pd.read_csv("assignment.csv")
df.head()
```

```python
# Displaying the basic information about the Data Set
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
print("\nUnique Values in Each Column:\n", df.nunique())
print("\nMinimum Values:\n", df.min())
print("\nMaximum Values:\n", df.max())
```

```python
df["Frailty"] = df["Frailty"].map({"N": 0, "Y": 1})
df
```

```python
df.columns = df.columns.astype(str).str.strip()
print(df.columns)
```

```python
# Data analysis and visualization
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Height'], y=df['Weight'], data=df,hue='Weight', palette="coolwarm", fliersize=0, notch=True, showmeans=True)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Boxplot graph between Weight and Height')
plt.grid(True)
plt.show()
```
