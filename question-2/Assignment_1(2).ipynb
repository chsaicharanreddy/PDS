```python
# Importing libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data from the data set.
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# visualization-1:- Histogram of Scores
plt.figure(figsize=(8, 5))
sns.histplot(df["math score"], bins=30, kde=True, color="#6B6BFF")
plt.xlabel("Math Score")
plt.ylabel("count")
plt.title("HistogramPlot of Math Scores with count")
plt.show()
```

```python
# visualization-2:- Analyzing the writing score and reading score between female and male
plt.figure(figsize=(8,6))
sns.lineplot(x="reading score", y="writing score", data=df, hue="gender", style='gender', markers=True, alpha=0.8)
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.title("Line Plot for Reading Score Vs Writing Score")
plt.grid(True)
plt.legend(title="Gender")
plt.show()
```

```python
# visualization-3:- Analyzing the Maths Score With Gender
plt.figure(figsize=(8, 5))
sns.boxplot(x="gender", y="math score", data=df, hue="gender", palette="coolwarm", fliersize=0, notch=True, showmeans=True)
plt.xlabel("Gender")
plt.ylabel('Maths Score')
plt.title("BoxPlot of Maths Score With Gender")
plt.grid(True)
plt.show()
```

```python
# visualization-4:- Bar Graph for Maths Score Vs Parental Level of Education
plt.figure(figsize=(8,7))
sns.barplot(x='parental level of education',y='math score', data=df, hue='gender', palette='Set1',
            order= df['parental level of education'].value_counts().index, errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.xlabel('parental level of education')
plt.ylabel('Maths Score')
plt.title("Bar Graph for Maths Score Vs Parental Level of Education")
plt.legend(title='Gender')
plt.grid(axis='y')
plt.show()
```
